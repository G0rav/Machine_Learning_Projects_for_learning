```python
mumbai_data_1 = mumbai_data.drop([
    'sl no.','lattitude','longitude','dealer','posted','property photo','property video','society photo','society video','locality photo','locality video','description'
], axis ='columns')
mumbai_data_1.head()
```

```python
mumbai_data_1 = mumbai_data_1.replace({'floor': ['Ground','ried']}, 0)
```

```python
mumbai_data_1 = mumbai_data_1.replace({
    'bedrooms':'[A-Za-z]',
    'floor':'[A-Za-z]',
    'total floors':'[A-Za-z]'
},'',regex = True)
mumbai_data_1
```

```python
mumbai_data_1['floor'] = mumbai_data_1['floor'].astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
```

```python
mumbai_data_2 = mumbai_data_1.copy()
```

```python
mumbai_data_2.iloc[:,0] = enc.fit_transform(mumbai_data_2.society)
mumbai_data_2.head()
```

```python
mumbai_data_2.iloc[:,2] = enc.fit_transform(mumbai_data_2.location)
mumbai_data_2.head()
```

```python
type_dummies = pd.get_dummies(mumbai_data_2['type'], drop_first = True)
type_dummies
```

```python
furnishing_dummies = pd.get_dummies(mumbai_data_2['furnishing'], drop_first=True)
furnishing_dummies
```

```python
age_dummies = pd.get_dummies(mumbai_data_2['age'], drop_first=True)
age_dummies
```

```python
final_data = pd.concat([mumbai_data_2,type_dummies,furnishing_dummies,age_dummies], axis = 'columns')
final_data.head()
```

```python
final_data.drop(['type','furnishing','age'], axis = 'columns', inplace = True)
final_data
```
